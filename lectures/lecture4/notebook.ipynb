```python
# Importing dependencies
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('student-exam-score.csv')
df.head(6)
```

```python
# 1. (10pts) Create mappings for the “parental level of education” feature update your
# dataframe with the mappings

# parent_level_education = df["parental level of education"].unique()
# parent_level_education_map = {parent_level_education[i]: i for i in range(len(parent_level_education))}
# print(parent_level_education_map)
# df["parental level of education"] = df["parental level of education"].map(lambda x: parent_level_education_map[x])
# df
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
mapped_values = encoder.fit_transform(df["parental level of education"].values)
df["parental_level_of_education_encoded"] = mapped_values
df.head(6)
```

```python
# 2. (10pts) Create new training and test data sets from the original data (remember to drop values in the “math-score” column with missing values). 
# Use the following code to create the training and test data sets:
from sklearn.model_selection import train_test_split


df_dropped_na_math_score = df.dropna(subset=["math score"])
# math_score_for_prediction = np.array(df_dropped_na_math_score["math score"]).reshape(-1, 1)
# parent_level_education_for_prediction = np.array(df_dropped_na_math_score["parental_level_of_education_encoded"]).reshape(-1, 1)

math_score_for_prediction = df_dropped_na_math_score[["math score"]]
parent_level_education_for_prediction = df_dropped_na_math_score[["parental_level_of_education_encoded"]]

x_train, x_test, y_train, y_test = train_test_split(parent_level_education_for_prediction, math_score_for_prediction, test_size=0.2, random_state=42)
len(x_train), len(x_test), len(y_train), len(y_test)
```

```python
# 3. (20pts) Using Linear Regression with “math-score” as the dependent variable (y) and “parental level of education” as the independent variable (x), fit the model with your training data and score your model with your test data. Note the score in your answer.
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)

print('Training score: {}'.format(model.score(x_train, y_train)))
print('Test score: {}'.format(model.score(x_test, y_test)))
```

```python
# 4. (20pts) Using Linear Regression with L1 Regularization, using “math-score” as the
# dependent variable (y) and “parental level of education” as the independent variable
# (x), fit the model with each alpha with your training data and score your model with
# your test data: .1, .15, .3, .6, .9, 1. Note the scores for each alpha in your answer

from sklearn.linear_model import Lasso

x_train, x_test, y_train, y_test = train_test_split(parent_level_education_for_prediction, math_score_for_prediction, test_size=0.2, random_state=42)


for alpha in [.1, .15, .3, .6, .9, 1]:
    model = Lasso(alpha=alpha)
    model.fit(x_train, y_train)
    print('Alpha: {}'.format(alpha))
    print('Training score: {}'.format(model.score(x_train, y_train)))
    print('Test score: {}'.format(model.score(x_test, y_test)))
    print()
```

```python
# 5.	(20pts) Using Linear Regression with L2 Regularization, using “math-score” as the dependent variable (y) and 
# “parental level of education” as the independent variable (x), fit the model with each alpha with your training 
# data and score your model with your test data: .1, .15, .3, .6, .9, 1. Note the scores for each alpha in your answer

from sklearn.linear_model import Ridge

x_train, x_test, y_train, y_test = train_test_split(parent_level_education_for_prediction, math_score_for_prediction, test_size=0.2, random_state=42)

for alpha in [.1, .15, .3, .6, .9, 1]:
    model = Ridge(alpha=alpha)
    model.fit(x_train, y_train)
    print('Alpha: {}'.format(alpha))
    print('Training score: {}'.format(model.score(x_train, y_train)))
    print('Test score: {}'.format(model.score(x_test, y_test)))
    print()
```

```python
# 7.	(30pts) Perform steps 3-5 using the features “math-score” and “reading-score” but divide up you training and test data differently.
print("-----------For math score and reading score-----------\n")
df_dropped_na_math_score = df.dropna(subset=["math score"])
math_score_for_prediction = df_dropped_na_math_score[["math score"]]
reading_score_for_prediction = df_dropped_na_math_score[["reading score"]]

model_details = []

for test_size in [0.1, 0.2, 0.3]:
  print('-----------Test size: {}-----------'.format(test_size))
  x_train, x_test, y_train, y_test = train_test_split(reading_score_for_prediction, math_score_for_prediction, test_size=test_size, random_state=42)
  print('-----------Linear Regression-----------')
  model = LinearRegression()
  model.fit(x_train, y_train)
  training_score = model.score(x_train, y_train)
  test_score = model.score(x_test, y_test)
  print('Training score: {}'.format(training_score))
  print('Test score: {}'.format(test_score))
  # append the model details to the list
  model_details.append({'model': 'Linear Regression', 'test_size': test_size, 'training_score': training_score, 'test_score': test_score})
  print()

  print('-----------Lasso - L1 Regularization-----------')
  for alpha in [.1, .15, .3, .6, .9, 1]:
      model = Lasso(alpha=alpha)
      model.fit(x_train, y_train)
      print('Alpha: {}'.format(alpha))
      training_score = model.score(x_train, y_train)
      test_score = model.score(x_test, y_test)
      print('Training score: {}'.format(training_score))
      print('Test score: {}'.format(test_score))
      model_details.append({'model': 'Lasso - L1 Regularization', 'test_size': test_size, 'alpha': alpha, 'training_score': training_score, 'test_score': test_score})
      print()

  print('-----------Ridge - L2 Regularization-----------')
  for alpha in [.1, .15, .3, .6, .9, 1]:
      model = Ridge(alpha=alpha)
      model.fit(x_train, y_train)
      print('Alpha: {}'.format(alpha))
      training_score = model.score(x_train, y_train)
      test_score = model.score(x_test, y_test)
      print('Training score: {}'.format(training_score))
      print('Test score: {}'.format(test_score))
      model_details.append({'model': 'Ridge - L2 Regularization', 'test_size': test_size, 'alpha': alpha, 'training_score': training_score, 'test_score': test_score})
      print()
```
